## <font color='blue'> <b>filter() & where()</b><font color='black'>

**filter()** fonksiyonu, sütun veya satır etiketlerine göre veri çerçevesinin ya da serinin bir alt kümesini almak için kullanılır. Bu, belirli sütunları ya da satırları seçmek istediğinizde kullanışlıdır.

**where()** fonksiyonu, bir veri çerçevesinde ya da seride belirli bir koşulu karşılayan değerleri korumak için kullanılır. Koşulu karşılamayan değerler ya NaN (varsayılan) olur veya isteğe bağlı olarak başka bir değerle değiştirilebilir.

[Pandas Official Documentation](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.filter.html),

[SOURCE02](https://appdividend.com/2020/03/19/pandas-filter-pandas-dataframe-filter-in-python-example/),
[SOURCE03](https://www.geeksforgeeks.org/python-pandas-dataframe-where/)
    




```python
import numpy as np
import pandas as pd 
import seaborn as sns
```

```python
data = {'Company':['Amazon', 'Amazon', 'OpenAI', 'OpenAI', 'Amazon', 'OpenAI', 'Amazon', 'OpenAI'],
        'Department':['IT', 'HR', 'IT', 'HR', 'IT', 'HR', 'HR', 'IT'],
        'Person':['Ahmet', 'Hatice', 'Fatma', 'Mehmet', 'Zeynep', 'Ali', 'Yusuf', 'Elif'],
        'Age':[32, 26, 35, 40, 25, 38, 30, 28],
        'Sales':[180, 160, 150, 180, 240, 200, 220, 160]}
```

```python
df =pd.DataFrame(data)
df
```

```python
df.filter(["Company", "Age"])
```

```python
df.filter(like= "art"  , axis=1)
```

```python
df.groupby("Company").filter(lambda x: x["Age"].mean() < 30)
```

```python
df.where(df.Age > 30)
```

```python
df.where(df.Age > 30, 0)
```
