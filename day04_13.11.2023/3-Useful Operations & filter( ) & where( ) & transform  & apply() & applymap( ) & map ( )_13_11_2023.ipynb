## <font color='blue'> <b>filter() & where()</b><font color='black'>

**filter()** fonksiyonu, stun veya satr etiketlerine gre veri erevesinin ya da serinin bir alt kmesini almak iin kullanlr. Bu, belirli stunlar ya da satrlar semek istediinizde kullanldr.

**where()** fonksiyonu, bir veri erevesinde ya da seride belirli bir koulu karlayan deerleri korumak iin kullanlr. Koulu karlamayan deerler ya NaN (varsaylan) olur veya istee bal olarak baka bir deerle deitirilebilir.

```python
import numpy as np
import pandas as pd 
import seaborn as sns
```

```python
data = {'Company':['Amazon', 'Amazon', 'OpenAI', 'OpenAI', 'Amazon', 'OpenAI', 'Amazon', 'OpenAI'],
        'Department':['IT', 'HR', 'IT', 'HR', 'IT', 'HR', 'HR', 'IT'],
        'Person':['Ahmet', 'Hatice', 'Fatma', 'Mehmet', 'Zeynep', 'Ali', 'Yusuf', 'Elif'],
        'Age':[32, 26, 35, 40, 25, 38, 30, 28],
        'Sales':[180, 160, 150, 180, 240, 200, 220, 160]}
```

```python
df =pd.DataFrame(data)
df
```

```python
df.filter(["Company", "Age"])
```

```python
df.filter(like= "art"  , axis=1)
```

```python
df.groupby("Company").filter(lambda x: x["Age"].mean() < 30)
```

```python
df.where(df.Age > 30)
```

```python
df.where(df.Age > 30, 0)
```
