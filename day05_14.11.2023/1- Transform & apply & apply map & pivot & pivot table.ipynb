## <font color='blue'> <b>transform()</b><font color='black'>
    
transform(), bir pandas DataFrame ya da Serisi'ndeki verileri deitirmek veya dntrmek iin kullanlan bir yntemdir. 

Belirtilen ilevi uygular ve sonucu yeni bir Seri veya DataFrame olarak dndrr.

zellikle gruplama ilemleri ile birlikte kullanldnda olduka yararldr.

[SOUREC01](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.transform.html) & [SOURCE02](https://www.analyticsvidhya.com/blog/2020/03/understanding-transform-function-python/)
    


 

```python
data = {'Company':['Amazon', 'Amazon', 'OpenAI', 'OpenAI', 'Amazon', 'OpenAI', 'Amazon', 'OpenAI'],
        'Department':['IT', 'HR', 'IT', 'HR', 'IT', 'HR', 'HR', 'IT'],
        'Person':['Ahmet', 'Hatice', 'Fatma', 'Mehmet', 'Zeynep', 'Ali', 'Yusuf', 'Elif'],
        'Age':[32, 26, 35, 40, 25, 38, 30, 28],
        'Sales':[180, 160, 150, 180, 240, 200, 220, 160]}
```

```python
df = pd.DataFrame(data)
df
```

 - transform verileri istedigimiz degere gore donusturur.<br>
 - DataFramenin uzunlugu kadar uzunluk verir.<br>

```python
df_num = df.select_dtypes(include= "number")
df_num
```

```python
df_num.Sales.transform(np.sqrt) # herbir degerin karekokunu aldi 
```

```python
df.groupby("Company")["Age"].transform("mean")# sirket icinde calisanlarin yaslarinin ortalamasini aldik
```

## <font color='blue'> <b>pivot() vs pivot_table()</b><font color='black'>
pivot(): verileri bir DataFrame'den baka bir DataFrame'e dntrr ve verileri bir stuna gre gruplar. Bu yntem, verileri yeni bir yapya dntrmek iin kullanlrken, herhangi bir toplama veya gruplama ilevi gerekletirmez. Duplicated deerler olduunda kullanlamaz.

pivot_table(): bir veya daha fazla stuna gre gruplama ilemi gerekletirir ve sonu olarak bir zet tablosu oluturur. Bu yntem, verilerin toplama, ortalama, medyan, minimum, maksimum, sayma gibi ilemlerle gruplanmas ve zetlenmesi iin kullanlr. Duplicated deerler olduunda kullanldr.

[SOURCE01](https://www.quora.com/In-pandas-whats-the-difference-between-dataframe-pivot-and-dataframe-pivot_table),
[SOURCE02](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.pivot.html),
[SOURCE03](https://www.w3resource.com/pandas/dataframe/dataframe-pivot.php),
[SOURCE04](https://jakevdp.github.io/PythonDataScienceHandbook/03.09-pivot-tables.html),
[SOURCE05](https://www.analyticsvidhya.com/blog/2020/03/pivot-table-pandas-python/),
[SOURCE06](https://pandas.pydata.org/docs/reference/api/pandas.pivot_table.html)  
    


    
    

```python
data = {'gender':['male', 'female', 'female', 'male', 'female', 'male'],
        'sport':['tennis', 'tennis', 'basketball', 'football', 'voleyball', 'basketball'],
        'status':["professional","professional","professional","amateur","amateur","amateur"],
        'age':[20, 24, 26, 23, 22, 21],
        'height':[185, 172, 175, 178, 182, 196],
        'weight':[83, 58, 62, 80, 65, 90]}

df = pd.DataFrame(data)
df
```

```python
df.pivot(index="gender", columns="sport", values=["age"])
```

```python
df.pivot(index="gender", columns="sport", values=["age", "height", "weight"])
```

```python
df.pivot(index="gender", columns="sport", values=["status"])
```

```python
# value parametresi icin kategorik degerlerimiz varsa,sadece "pivot" fonk.kullanabiliriz.

# hata verir.

df.pivot_table(index="gender", columns="sport", values=["status"])
```

```python
df.loc[2, "sport"] = "tennis"
df
```

```python
df.pivot_table(index="gender", columns="sport", values=["age"]) # default = mean 
```

```python
df.pivot_table(index="gender", columns="sport", values=["age"],aggfunc ="mean")
```
