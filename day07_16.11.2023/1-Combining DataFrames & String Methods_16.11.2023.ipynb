## <font color='green'> <b>concat()</b><font color='black'>

**concat()**, iki veya daha fazla DataFrame'i satrlar veya stunlar baznda birletirmek iin kullanlr.

Farkl veri kmelerini ayn veri kmesinde bir araya getirmek veya veri kmesine yeni stunlar veya satrlar eklemek iin kullanlabilir.

Veri kmeleri arasnda herhangi bir ortak stun gerekmez.

[Source 02](https://sparkbyexamples.com/pandas/pandas-concat-dataframes-explained/),

```python
pd.concat([one, two], ignore_index=True) 
```

```python
pd.concat([one, two], axis=1) # 
```

```python
pd.concat([one, two], axis=1, ignore_index=True)
```

## <font color='green'> <b>merge()</b><font color='black'>
    
**merge()**, iki veya daha fazla veri kmesini, ortak bir stuna gre birletirme ilemidir. 

Ortak stunlardaki deerlere gre veriler birletirilir ve tek bir veri kmesi oluturulur. 

merge(), veri kmesindeki eksik veya farkl deerleri tamamlamak veya farkl veri kaynaklarn bir araya getirmek iin ska kullanlr.

[Source 01](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.merge.html), 
[Source 02](https://www.w3schools.com/python/pandas/ref_df_merge.asp), 
[Source 03](https://www.digitalocean.com/community/tutorials/pandas-merge-two-dataframe),
[Source 04](https://towardsdatascience.com/why-and-how-to-use-merge-with-pandas-in-python-548600f7e738)





```python
x = pd.DataFrame({'key': ['K0', 'K1', 'K2'],
                     'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']})

y = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
x
```

```python
y
```

```python
pd.merge(x, y, how="inner", on="key") # ortak sutunlar tekrar etmiyor 
```

```python
pd.merge(x, y, how="outer")
```

```python
pd.concat([x,y])  # satirlar ve sutunlar bazinda birlestirdik.
```

```python
pd.merge(x, y, how="left", on="key")
```

```python
pd.merge(x, y, how="right", on="key")
```

```python
df1 = pd.DataFrame({'lkey': ['x', 'y', 'z', 'x'],
                    'lvalue': [2, 3, 5, 7]})

df2 = pd.DataFrame({'rkey': ['a', 'x', 'z', 'b'],
                    'rvalue': [7, 8, 9, 10]})
```

```python
df1
```

```python
df2
```

```python
# pd.merge(df1, df2, how="inner") # hata verir ortak bir sutun yoktur
```

```python
pd.merge(df1, df2, left_on="lkey", right_on="rkey")  # default = inner
```

## <font color='green'> <b>join()</b><font color='black'> 
    
**join()**, iki veya daha fazla DataFrame'i belirli bir stuna gre birletirme ilemidir.

Daha ok SQL tabanl veri ileme ilemlerinde kullanlr. 

[Source 01](https://sparkbyexamples.com/pandas/pandas-join-explained-with-examples/), 
[Source 02](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.join.html), 
[Source 03](https://www.w3schools.com/python/pandas/ref_df_join.asp) 
    
 

```python
df3 = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']},
                      index = ['K0', 'K1', 'K2']) 

df4 = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                    'D': ['D0', 'D2', 'D3']},
                      index = ['K0', 'K2', 'K3'])
```

```python
df3
```

```python
df4
```

```python
df3.join(df4, how="inner")
```

```python
df4.join(df3) # join yanyana alip birlestirir.
```

```python
df5 = pd.DataFrame({'key': ['K0', 'K2', 'K3', 'K4', 'K5', 'K6'],
                   'X': ['X0', 'X2', 'X3', 'X4', 'X5', 'X6']})

df6 = pd.DataFrame({'key': ['K0', 'K2', 'K3'],
                      'Y': ['Y0', 'Y2', 'Y3']})
```

```python
df5
```

```python
df6
```

```python
# hata verir  df5.join(df6) #
```

```python
df5.join(df6, lsuffix="_left", rsuffix="_right")  # stunlarn isimlerinin farkl olmas gerekiyor
                                                  #  Isuffix ayni sutunlarin sonuna yazdigimiz ekleri koyar
#lsuffix =soldan ekle
#rsuffix = sagdan ekle 
```

```python
df5
```

```python
df6
```

```python
df5.set_index("key") # 
```

```python
df6.set_index("key")
```

```python
df5.set_index("key").join(df6.set_index("key"))
```

```python
df5.set_index("key").join(df6.set_index("key")).reset_index()
```
